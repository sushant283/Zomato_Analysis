```python
#IMPORTING LIB
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

# Upload a CSV file
uploaded = files.upload()
#IMPORTING DATA SET
dataframe=pd.read_csv("Zomato data .csv")
dataframe
```

```python
#DATA PREPROCCESSING
#Handling the rating value and simplifying it
def handlerate(value):
    value=str(value).split('/')
    value=value[0];
    return float (value)

dataframe['rate']=dataframe['rate'].apply(handlerate)
dataframe
```

```python
dataframe.head()
```

```python
#Finding null values
dataframe.isnull().sum()
```

```python
#TYPE OF RESTURANT THERE

category_counts = dataframe['listed_in(type)'].value_counts()
plt.figure(figsize=(10,6))
plt.bar(category_counts.index, category_counts.values)
plt.xlabel('Category')
plt.ylabel('Number of Listings')
plt.title('Number of Times Each Category is Listed')
plt.show()
```

```python
#WHICH TYPE OF RESTURANT GETS MAX VOTES

grpdata=dataframe.groupby('listed_in(type)')['votes'].sum()
res=pd.DataFrame({'votes':grpdata})
plt.plot(res,c='red',marker="o")
plt.xlabel("Type of Resturant ")
plt.ylabel("Votes")
plt.title('WHICH TYPE OF RESTURANT GETS MAX VOTES')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="ticks")
plt.figure(figsize=(8, 5))
plt.hist(dataframe['rate'], bins=15, color='lightblue', edgecolor='black')
plt.xlabel("Ratings")
plt.ylabel("Frequency")
plt.title("Restaurant Rating Distribution")
plt.show()
```

```python
#Average order spending by couples
category_counts = dataframe['approx_cost(for two people)'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=category_counts.index, y=category_counts.values)
plt.xlabel('Cost (for two people)')
plt.ylabel('Counts')
plt.title('Average Order Spending by Couples')
plt.xticks(rotation=45)
plt.show()
```

```python
#no of online orders
category_counts = dataframe['online_order'].value_counts()
plt.bar(category_counts.index, category_counts.values,color='yellow')
plt.ylabel('Order')
plt.title('Number of online order')
plt.show()
```
